# **Existe associação entre as variávies qualitativas e as duas doenças?**


## 1) Primeiro, vamos analisar a independência entre a ocorrência de náuseas e a Inflamação da Bexiga.

```python
# Tabela de dupla entrada
tabela_dupla = pd.crosstab(index=df['Ocorrência de náusea'], columns=df['Inflamação da bexiga urinária'])
tabela_dupla
```

```python
from scipy.stats import chi2_contingency
res = chi2_contingency(tabela_dupla)
estatistica = res.statistic
p_value = res.pvalue
print("A tabela de valores esperados: ", res.expected_freq)
print("O valor da estatística do qui-quadrado é = ", estatistica)
print("O valor do p-value é: ", p_value)
```

#Conclusão?
## Sobre a independência entre a ocorrência de náuseas e a Inflamação da Bexiga, descartamos a hipótese de independencia, sim ou não?

# Como o p-value é maior que 5%, nós NÃO descartamos a hipótese de indenpendência entre as varíaveis, logo, NÃO PODEMOS AFIRMAR QUE HÁ dependencia ou causalidade entre a ocorrência de náuseas e a inflamação da Bexiga.

```python
import pandas as pd
df2 = pd.read_excel("/content/Conjunto de dados de registros clínicos de insuficiência cardíaca.xlsx")
df2
```

```python
import matplotlib.pyplot as plt

# Dados do exemplo
x=df2["creatinine_phosphokinase"]
y=df2['serum_sodium']

# Plotagem do gráfico de dispersão
plt.scatter(x, y)
plt.title('Gráfico de dispersão')
plt.xlabel('x: CPK - creatinina fosfoquinase')
plt.ylabel('y: Sódio')
plt.show()
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from scipy.stats import f

# Cálculo das médias
mean_y = np.mean(y)
mean_x = np.mean(x)

# Cálculo da soma de quadrados total (SQT)
SQT = np.sum((y - mean_y)**2)

# Ajuste do modelo de regressão linear simples
x = sm.add_constant(x)
model = sm.OLS(y, x).fit()

# Coeficientes da reta ajustada
b0 = model.params[0]
b1 = model.params[1]

# Cálculo da soma de quadrados explicada (SSE)
y_pred = model.predict(x)
SQReg = np.sum((y_pred - mean_y)**2)

# Cálculo da soma de quadrados não explicada (SSR)
SQRes = SQT - SQReg

# Cálculo dos graus de liberdade
n = len(y)
df_model = 1
df_residual = n - 2

# Cálculo dos quadrados médios
QMT = SQT/(n-1)
QMReg = SQReg / df_model
QMRes = SQRes / df_residual

R_quadrado = SQReg/SQT

# Cálculo do valor F e do p-valor
f_value = QMReg / QMRes
p_value = 1 - f.cdf(f_value, df_model, df_residual)

# Imprimir Conclusão
if p_value < 0.05:
    print("Há relação de dependência linear entre as duas variáveis quantitativas. \nPortanto, uma reta de regressão linear pôde ser ajustada. \n")
else:
    print("NÃO há relação de dependência linear entre as duas variáveis quantitativas. \nPortanto, uma reta de regressão linear NÃO pôde ser ajustada. \n")


# Impressão dos resultados
print('Coeficente linear da reta estimada', b0)
print('Coeficente angular da reta estimada', b1)
print('Soma de quadrados total (SST):', SQT)
print('Soma de quadrados explicada (SSE):', SQReg)
print('Soma de quadrados não explicada (SSR):', SQRes)
print('Quadrado médio explicado (MSE):', QMReg)
print('Quadrado médio não explicado (MSR):', QMRes)
print('Valor F:', f_value)
print('p-valor:', p_value)
print('R_quadrado:', R_quadrado)

```
