```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import tensorflow as tf
from sklearn.neural_network import MLPClassifier
```


## 1. Leia a Base de Dados `diabetes.csv`


```python
dados_diabetes = pd.read_csv('diabete.csv')
```


## 2. Análise Inicial da Base de Dados


## 3. Separação dos Dados
Separe os dados em:
- **X**: Atributos
- **y**: Classe alvo (coluna indicando se tem (1) ou não (0) diabetes "Diabetes_binary")


```python
X = dados_diabetes.drop("Diabetes_binary", axis=1)

y = dados_diabetes["Diabetes_binary"]
```


## 4. Divisão dos Dados
Separe os dados em:
- **Treinamento**: 60%
- **Validação**: 20%
- **Teste**: 20%


```python
X_treino_temp, X_teste, y_treino_temp, y_teste = train_test_split(X, y, test_size=0.4, random_state=42)

X_treino, X_validacao, y_treino, y_validacao = train_test_split(X_treino_temp, y_treino_temp, test_size=0.5, random_state=42)

print(f"Conjunto de Treinamento: {len(X_treino)} amostras")
print(f"Conjunto de Validação: {len(X_validacao)} amostras")
print(f"Conjunto de Teste: {len(X_teste)} amostras")
```


## 5. Normalização dos Dados (Apenas X)
- **a.** Fit transform nos dados de treinamento
- **b.** Transform nos dados de teste
- **c.** Transform nos dados de validação


```python
scaler = StandardScaler()
X_treino_norm = scaler.fit_transform(X_treino)

X_teste_norm = scaler.transform(X_teste)

X_validacao_norm = scaler.transform(X_validacao)
```


## 6. Treinamento de Modelos SVM
Treine 4 modelos de SVM com diferentes parâmetros para o "C", kernel ou gama. Nomeie cada modelo especificamente.
- **Documentação da biblioteca**: [Scikit-learn SVM](https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html)


```python
modelo_svm_padrao = SVC(random_state=42)
modelo_svm_padrao.fit(X_treino_norm, y_treino)
y_pred_padrao = modelo_svm_padrao.predict(X_teste_norm)
acuracia_padrao = accuracy_score(y_teste, y_pred_padrao)

modelo_svm_poly = SVC(kernel='poly', random_state=42)
modelo_svm_poly.fit(X_treino_norm, y_treino)
y_pred_poly = modelo_svm_poly.predict(X_teste_norm)
acuracia_poly = accuracy_score(y_teste, y_pred_poly)

modelo_svm_rbf = SVC(kernel='rbf', random_state=42)
modelo_svm_rbf.fit(X_treino_norm, y_treino)
y_pred_rbf = modelo_svm_rbf.predict(X_teste_norm)
acuracia_rbf = accuracy_score(y_teste, y_pred_rbf)

modelo_svm_custom = SVC(C=0.1, kernel='linear', random_state=42)
modelo_svm_custom.fit(X_treino_norm, y_treino)
y_pred_custom = modelo_svm_custom.predict(X_teste_norm)
acuracia_custom = accuracy_score(y_teste, y_pred_custom)
```


## 7. Treinamento de Redes Neurais
Treine 3 redes neurais com diferentes estruturas por 20 épocas. Nomeie cada modelo especificamente.


```python
modelo_rn_1 = MLPClassifier(hidden_layer_sizes=(100,), max_iter=20, random_state=42)
modelo_rn_1.fit(X_treino_norm, y_treino)
y_pred_rn_1 = modelo_rn_1.predict(X_teste_norm)
acuracia_rn_1 = accuracy_score(y_teste, y_pred_rn_1)

modelo_rn_2 = MLPClassifier(hidden_layer_sizes=(50, 50), max_iter=20, random_state=42)
modelo_rn_2.fit(X_treino_norm, y_treino)
y_pred_rn_2 = modelo_rn_2.predict(X_teste_norm)
acuracia_rn_2 = accuracy_score(y_teste, y_pred_rn_2)

modelo_rn_3 = MLPClassifier(hidden_layer_sizes=(30, 20, 10), max_iter=20, random_state=42)
modelo_rn_3.fit(X_treino_norm, y_treino)
y_pred_rn_3 = modelo_rn_3.predict(X_teste_norm)
acuracia_rn_3 = accuracy_score(y_teste, y_pred_rn_3)
```


## 8. Avaliação dos Modelos
Avalie todos os 7 modelos treinados.
